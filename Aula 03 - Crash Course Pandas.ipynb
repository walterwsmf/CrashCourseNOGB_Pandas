### Definindo grupo 

````python
g = data.groupby(by=['Sex','Pclass'])
````

````python
g,type(g)
````

````python
g.describe()
````

````python
g.ngroups
````

````python
def CV(group):
    return {
        'media': np.mean(group['Fare']),
        'std': np.std(group['Fare']),
        'mediana': np.median(group['Fare'].dropna()),
        'CV': np.mean(group['Fare'])/np.std(group['Fare']),
    }
````

````python
g.apply(CV)
````

````python
g.apply(CV).apply(pd.Series)
````

````python
g.pipe(lambda grp: grp['Fare'].mean()/grp['Fare'].std())
````

````python
g['Fare'].transform(lambda x: x.mean()/x.std())
````

````python
data['CV Fare'] = g['Fare'].transform(lambda x: x.mean()/x.std())
````

````python
data.head()
````

````python
data.apply(lambda df: (df['Name'])+' '+str(df['Cabin']), axis=1)
````

## Queremos incluir o resultado da análise do groupby na tabela original:

- Usar concat

- Usar o merge

O merge funciona como a ideia de JOIN de sql, dado uma coluna semelhante em ambas as tabelas, podemos unir ambas pelos elementos semelhantes entre elas:

````python
g = data.groupby(by=['Sex','Pclass'])

g = g.apply(CV).apply(pd.Series).reset_index()

g
````

````python
data.isna().sum()
````

As variaveis **g** e **data** possuem as colunas **Sex** e **Pclass** semelhantes:

````python
pd.merge(data, g, left_on=['Sex','Pclass'], right_on=['Sex','Pclass'],  how='inner')
````

````python
# pd.merge(data, g, on=['Sex','Pclass'], how='inner')
````

````python
help(pd.merge)
````

A rotina merge também está inserida dentro do proprio dataframe:

````python
data.merge(g, right_on=['Sex','Pclass'], left_on=['Sex','Pclass'], how='inner')
````

````python
data = data.merge(g, right_on=['Sex','Pclass'], left_on=['Sex','Pclass'], how='inner')
````

````python
data.head()
````

````python
data.sort_values(by=['Sex','Pclass'],inplace=True)
````

````python
data.head()
````

### colunas semelhantes em ambas as bases que nao sao usadas no merge

````python
data.merge(g, 
           right_on=['Sex','Pclass'], 
           left_on=['Sex','Pclass'], 
           how='inner'
          ).head()
````

````python
data.merge(g, 
           right_on=['Sex','Pclass'], 
           left_on=['Sex','Pclass'], 
           how='inner', 
           suffixes=('_df_right','_df_letf')
          ).head()

#suffixes=(r'$\alpha$','_df_letf')
````

Podemos incluir um sufixo ao final do nome das colunas dentro da rotina do merge, tal que, possamos identificar a coluna semelhante dada sua origem.

## se quisermos, podemos fornecer uma lista com valores a serem utilizados de nivel:

````python
df = data.dropna(subset=['Age','Fare'])

C, xedges, yedges = np.histogram2d(df['Age'],df['Fare'])

C, xedges, yedges
````

````python
CT = C.T
CT
````

Vamos agora normalizar pelo maximo, pois os levels tem que ser entre 0 e 1.
E usando a regra de 68–95–99.7:

````python
CT = CT/CT.max()
CT
````

````python
levels = [np.percentile(CT, 68), np.percentile(CT, 95), np.percentile(CT, 99.7)]
levels
````

````python
CT.flatten()
````

````python
plt.figure(figsize=(14,6))
sns.ecdfplot(CT.flatten())

plt.axhline(0.68,color='r',ls='--')
plt.axhline(0.95,color='b',ls='--')
plt.axhline(0.997,color='g',ls='--')
plt.ylim(0,1.1)
````
